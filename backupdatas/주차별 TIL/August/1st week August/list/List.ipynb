### 버블 정렬(Bubble Sort)
인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식

#### 버블 정렬 과정
1. 첫 번째 원소부터 인접한 원소끼리 계속 자리를 교환하면서 맨 마지막 자리까지 이동한다.
2. 한 단계가 끝날 때 마다 남아있는 원소들 중 가장 큰 원소가 마지막 자리로 정렬된다

#### 버블 정렬 시간 복잡도
O(n^2)


````python
# 버블 정렬 알고리즘(오름차순 case)
def BubbleSort(arr):
    N = len(arr)
    for i in range(N-1, 0, -1): # N
        for j in range(i): # N
            if arr[j] > arr[j+1]: # 비교
                arr[j], arr[j+1] = arr[j+1], arr[j] # 교환
    return arr

# 정렬되지 않은 1차원 배열
arr = [3, -1, 0, 1, 9, 6, 2, 6]
print(BubbleSort(arr)) # 버블정렬 후 결과 반환
````

### 카운팅 정렬(Counting Sort)
항목들의 순서를 결정하기 위해 집합에 각 항목이 몇 개씩 있는지 세는 작업(counts)를 하여 선형 시간에 정렬하는 효율적인 알고리즘
- 정수/정수로 표현 가능한 자료에 대해서만 적용 가능
- 카운트들을 위한 공간을 할당하기 위해 가장 큰 정수를 알아야 함
- n이 비교적 작을 때 사용 가능(교환 X, 비교 O)


#### 카운팅 정렬 과정
1. counts에 배열의 최댓값 크기 + 1만큼의 공간을 할당한다(counts = [0] * max(array))
2. 배열의 데이터에 접근하여 해당 데이터를 index로 하는 counts에 데이터의 개수만큼 1을 더해준다
3. 각 항목의 앞에 위치하는 항목의 개수를 반영하기 위해(누적) counts의 원소를 조정한다
4. 정렬된 배열을 담을 새로운 배열(temp)을 데이터 크기만큼의 공간을 할당하여 생성한다(temp = [0] * len(array))
5. 배열의 끝 인덱스부터 역순으로 접근하여, temp에 순서대로 데이터들을 정렬하여 할당한다

#### 카운팅 정렬의 시간복잡도
O(n+k) : n은 배열의 크기, k는 정수의 최대값

````python
# 카운팅 정렬
def CountingSort(arr):
    N = len(arr)
    k = max(arr)
    
    counts = [0] * (k+1)
    temp = [0] * N

    # 데이터의 개수 파악(counts의 idx에 저장된 값 = 해당 idx를 값으로 하는 데이터의 개수)
    for ele in arr:
        counts[ele] += 1
    # 데이터의 개수 누적 파악
    for i in range(len(arr)-1):
        counts[i+1] += counts[i]
    
    # temp에 정렬된 배열 할당
    for i in range(N-1, -1, -1):
        # 누적된 개수 차감
        counts[arr[i]] -= 1
        # 데이터를 순서를 고려하여 temp에 할당
        temp[counts[arr[i]]] = arr[i]
    return temp
    
# 정렬되지 않은 1차원 배열
arr = [3, 0, 1, 0, 5, 3, 9, 6, 2, 6]
print(CountingSort(arr)) # 버블정렬 후 결과 반환
````

## 6. 탐욕 알고리즘

### 탐욕 알고리즘(Greedy Algorithm)
최적해를 구하는 데 사용되는 근시안적인 방법
- 하나를 결정할 때 마다 그 순간에 최적이라고 생각하는 것을 선택
- 각 선택의 시점에서 이루어지는 결정은 지역적으로는 최적일 수 있으나, 그것들을 수집하여 내린 최종 해답이 항상 최적이라는 보장은 X

#### 탐욕 알고리즘 과정
1. 해 선택 : 현재 상태에서 부분 문제의 최적 해를 구한 뒤, 이를 부분해 집합(Solution Set)에 추가
2. 실행 가능성 검사 : 새로운 부분해 집합이 실행 가능한지 확인

## 8. 부분집합 생성

### 부분집합 생성
원소가 N개일 때 부분집합의 수 : 2 ^ N(공집합 포함)

#### 비트 연산자

- 1 << N : 2^N(원소가 N개일 경우의 모든 부분집합의 수)
- i & (1 << j) : i의 j번째 비트가 1인지 아닌지 검사
```python
비트 연산자 사용법
arr = [3, 6, 7, 1, 5, 4]
N = len(arr)
subsets = []
for i in range(1 << N): # 부분집합의 개수만큼 반복
    subset = []
    for j in range(N): # 원소의 수만큼 비트 비교
        if i & (1 << j): # i의 j번 비트가 1인 경우, j번 원소를 부분집합의 원소로 넣기
            subset.append(arr[j])
    subsets.append(subset)
print(subsets)
```

````python
# 부분집합 만들기
arr = [3, 6, 7, 1, 5, 4]
N = len(arr)
subsets = []
for i in range(1 << N):
    subset = []
    for j in range(N):
        if i & (1 << j):
            subset.append(arr[j])
    subsets.append(subset)
print(subsets)
````

## 9. 검색

### 검색(Search)

#### 검색의 종류

### 순차 검색(Sequential Search)
배열이나 연결리스트 등의 순차구조로 구현된 자료구조에서 순서대로 순회하며 원하는 항목을 찾는 방법
- 가장 간단하고 직관적인 검색 방법
- 검색 대상의 수가 많은 경우 비효율정

#### 정렬되어있지 않은 경우
1. 첫 번째 원소부터 순서대로 검색 대상과 키 값이 같은 원소가 있는지 비교
2. 키 값이 동일한 원소를 찾을 경우, 그 원소의 인덱스 반환
3. 자료구조의 마지막에 이를 때 까지 검색 대상을 찾지 못하면 실패
- 찾고자 하는 원소의 순서에 따라 비교회수 결정 : 평균 비교회수 = (1/N) * (1+2+,,,+N) = (n+1)/2
- 시간 복잡도 : O(N)

#### 정렬되어있는 배열 순차 검색
자료를 순차적으로 검색하면서 키 값을 비교하여, 원소의 키 값이 검색 대상의 키 값보다 클 경우 검색 종료
- 찾고자 하는 원소의 순서에 따라 비교회수 결정 : 검색 실패시 정렬되어있지 않은 경우보다 평균 비교회수가 반으로 감소
- 시간 복잡도 : O(N)

````python
# 순차 검색
# 정렬되어있지 않은 경우
def SequentialSearch_1(arr, key):
    N = len(arr)
    i = -1
    while i < N and arr[i] != key:
        i += 1
        if i < N and arr[i] == key:
            return i
        
    return -1
# 정렬되어있는 경우
def SequentialSearch_2(arr, key):
    N = len(arr)
    i = -1
    while i < N and arr[i] < key:
        i += 1
        if i < N and arr[i] == key:
            return i
    return -1
````

### 이진 검색(Binary Search)
자료의 가운데에 있는 항목의 키 값과 비교하여 다음 검색의 위치를 결정하고 검색을 진행하는 방법
- 목적 키를 찾을 때 까지 반복 수행함으로써 검색 범위를 반으로 줄여가면서 보다 빠르게 검색 수행
- 자료가 정렬되어 있을 때 이진 검색 사용 가능

#### 이진 검색 과정
1. 자료의 중앙 원소를 고른다
2. 중앙 원소의 값과 찾고자 하는 목표 값을 비교한다
3. 목표 값이 중앙 원소의 값보다 작을 경우, 자료의 왼쪽 반에 대해 새로 검색을 진행한다. 목표값이 중앙 원소의 값보다 클 경우, 자료의 오른쪽 반에 대해 새로 검색을 진행한다.
4. 찾고자 하는 값을 찾을 때 까지 위의 과정을 반복한다

## 10. 선택 정렬

### 선택 정렬(Selection Sort)
주어진 자료들 중 가장 작은 값의 원소부터 차례대로 선택하여 위치를 교환하는 방식
- 셀렉션 알고리즘을 전체 자료에 적용

#### 선택 정렬 과정
1. 주어진 리스트에서 최소값을 찾는다
2. 최소값을 리스트의 맨 앞과 교환한다
3. 맨 처음 위치를 제외한 나머지를 대상으르 위의 과정을 반복한다
- 시간 복잡도 : O(N^2)

````python
# 선택 정렬 알고리즘
def SelectionSort(arr):
    N = len(arr)
    for i in range(N-1):
        min_idx = i
        for j in range(i+1, N):
            if arr[min_idx] > arr[j]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        
````
