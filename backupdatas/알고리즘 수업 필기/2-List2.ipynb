#### 비트 연산자
1. & : 비트단위로 AND 연산 시행
    - 각 자리 비트끼리 and 연산으로 인해 나온 값이 결과의 자리 비트(ex. 0 and 1 -> 0 / 1 and 1 -> 1)
2. | : 비트단위로 OR 연산 시행
    - 각 자리 비트끼리 and 연산으로 인해 나온 값이 결과의 자리 비트(ex. 0 and 1 -> 0 / 1 and 1 -> 1)
3. << : 피연산자의 비트 열을 왼쪽으로 이동 후 기존 자리는 0으로 채워넣음
   - ex. A = 1011 ---> A << 4 : 10110000(= A * (2 ** 4))
4. >> : 피연산자의 비트열을 오른쪽으로 이동
   - ex. A = 1011010000 --> A >> 4 : 101101(= A // (2 ** 4)) 
- <<와 >> 연산을 할 때 자리를 넘길 경우, 날라감!!!
++
- 1 << n == 1 * (2 ** n) == 2 ** n 즉, 원소가 n개인 집합의 모든 부분집합의 개수

#### 비트 연산자를 이용한 부분집합 생성 방법
i & (1 << j) : i의 j번째 비트가 1인지 아닌지 검사
- i : 내가 생성한 비트 열 / j : 검사할 비트 순서
- 변하는 길이에 대한 대응 가능

```python
arr = [3, 6, 7, 1, 5, 4]
n = len(arr)
# bit = [0, 0, 0, 0, 0, 0] 즉, 여섯 자리의 비트를 가지고 비교해야 함(= n)
for i in range(1 << n): # 모든 부분 집합의 개수에 대한 for 문 진행
    for j in range(n): # 집합의 개수만큼 비트 비교
        if i & (1 << j): # i의 j번 비트가 1일 때,
            print(arr[j], end = ',')

```

## 3. 검색

### 검색
저장되어 있는 자료 중에서 원하는 항목을 찾는 작업
- 목적하는 탐색 키(search key)를 가진 항목을 찾는 것
- 탐색 키 : 자료를 구별하여 인식할 수 있는 키

#### 검색의 종류
1. 순차 검색(sequential search)
2. 이진 검색(binary search)
3. 해쉬(hash)

#### 순차 검색
일렬로 되어 있는 자료를 순서대로 검색하는 방법
- 가장 간단하고 직관적
- 배열, 연결 리스트 등 순차구조로 구현된 자료구조에서 원하는 항목을 찾을 때 유용
- 검색 대상의 수가 많은 경우 수행시간이 급격히 증가하여 비효율적(모든 index에 접근)

#### 정렬되어 있지 않은 경우 순차 검색
시간 복잡도 : (1/n) * (1 + 2 + 3 + ... + n) = (n+1)/2 = O(n)
1. 첫 번째 원소부터 순서대로 검색 대상과 키 값이 같은 원소가 있는지 비교하며 찾기
2. 키 값이 동일한 원소를 찾으면 그 원소의 인덱스 반환
3. 자료구조의 마지막에 이를 때 까지 검색 대상을 찾지 못하면 검색 실패

#### 정렬되어 있는 경우 순차 검색
- 시간 복잡도 : O(n)
오름차순이라 가정
- 순차적으로 검색하며 키 값을 비교하여, 원소의 키 값이 검색 대상의 키 값보다 크면 찾는 원소가 없기 때문에 더 이상 검색하지 않고 검색을 종료
- 실패하는 경우의 평균 비교 회수가 정렬되지 않는 경우보다 반으로 줄어듦

### 이진 검색
자료의 가운데에 있는 항목의 키 값과 비교하여 다음 검색의 위치를 결정하고 검색을 진행하는 방법
- 검색 범위를 반으로 줄여가면서 빠르게 검색을 수행
- 자료가 정렬되어 있을 때 사용 가능한 검색 방법

#### 이진검색 과정
1. 자료의 중앙에 있는 원소를 고른다
2. 중앙의 원소 값과 목표 값을 비교한다
3. 목표값이 중앙 값보다 작을 경우, 중앙 값 기준 왼쪽 데이터에 대해서 검색을 수행한다. 중앙 값보다 클 경우, 중앙 값 기준 오른쪽 데이터에 대해서 검색을 수행한다
4. 찾고자 하는 값을 찾을 때 까지 위의 과정을 반복한다
- 검색 범위의 시작점(min_idx)과 종료점(max_idx)을 이용하여 검색을 반복 수행한다.
- 자료의 삽입, 삭제 등 자료구조 내의 수정이 발생한 경우 배열의 상태를 항상 정렬 상태로 유지하는 작업이 필요하다
```python
# 이진검색 함수
def BinarySearch(arr, N, key):
    # 시작 인덱스와 끝 인덱스 초기화
    start = 0
    end = N-1

    # 시작 인덱스가 끝 인덱스를 넘어갈 때 까지 반복
    while start <= end:
        # 검색을 진행할 범위의 중간 인덱스 생성
        middle = (start + end) // 2
        # 중앙 인덱스의 값이 찾고자 하는 값과 같을 경우 검색 종료
        if arr[middle] == key:
            return True
        # 찾고자 하는 값보다 클 경우 왼쪽 구간을 선택
        elif arr[middle] > key:
            end = middle - 1
        else:
            end = start + 1
    # while문에서 빠져나온 경우 즉, 찾고자 하는 키가 해당 배열에 없는 경우(검색 실패)
    return False

# 재귀함수를 이용한 이진 검색 함수
def BinarySearch_2(arr, low, high, key):
    if low > high:
        return False
    else:
        middle = (low + high) // 2
        if key == arr[middle]:
            return True
        # 키 값이 더 오른쪽에 있을 경우
        elif key > arr[middle]:
            # 새로운 low 할당하여 호출
            return BinarySearch(arr, middle + 1, high, key)
        # 키 값이 더 왼쪽에 있을 경우
        elif key < arr[middle]:
            return BinarySearch(arr, low, middle - 1, key)
        

### 선택 정렬
주어진 자료들 중 가장 작은 값의 원소부터 차례대로 선택하여 위치를 교환하는 방식
- 셀렉션 알고리즘을 전체 자료에 적용
- 인덱스를 이용하여 최소값의 위치를 파악(해당 인덱스의 값을 따로 저장 X, index로만)
- 시간 복잡도 : O(n^2)
#### 선택 정렬 과정

2. 그 값을 리스트의 맨 앞에 위치한 값과 교환한다
3. 맨 처음 위치를 제외한 나머지 리스트를 대상으로 위의 과정을 반복한다
