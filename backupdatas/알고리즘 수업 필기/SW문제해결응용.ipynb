## 1. SW 문제 해결

### SW 문제 해결

#### 학습목표
- SW 문제해결 역량이 무엇인지 이해하고, 역량을 강화하는 방법 이해
- 효율적인 알고리즘의 필요성을 이해하고 알고리즘의 성능 측정 방법 중 하나인 시간복잡도에 대해 이해
- 프로그램을 작성하기 위한 기본 중 표준 입출력 방법에 대해 이해
- 비트 수준의 연산과 알고리즘에 대해 이해
- 컴퓨터에서의 실수 표현 방법에 대해 이해

#### 프로그래밍 제약 조건과 요구사항
- 프로그래밍 언어의 특성
- 프로그램이 동작할 HW과 OS에 대한 지식
- 라이브러리들의 유의사항들
- 프로그램이 사용할 수 있는 최대 메모리
- 사용자 대응 시간 제한
- 재사용성이 높은 간결한 코드

#### SW 문제 해결 역량이란?
프로그램을 하기 위한 많은 제약 조건과 요구사항을 이해하고 최선의 방법을 찾는 능력
- 프로그래머가 사용하는 언어, 라이브러리, 자료구조, 알고리즘에 대한 지식을 연결하여 큰 그림을 만드는 능력

#### SW 문제 해결 과정
1. 문제를 읽고 이해한다
2. 문제를 익숙한 용어로 재정의한다
3. 어떻게 해결할지 계획을 세운다
4. 계획을 검증한다
5. 프로그램으로 구현한다
6. 어떻게 풀었는지 돌아보고, 개선할 방법이 있는지 찾아본다

## 2. 복잡도 분석

### 알고리즘
유한한 단계를 통해 문제를 해결하기 위한 절차나 방법
- 컴퓨터가 어떤 일을 수행하기 위한 단계적 방법

#### 알고리즘의 효율
- 공간적 효율성 : 연산량 대비 얼마나 적은 메모리를 요하는지
- 시간적 효율성 : 연산량 대비 얼마나 적은 시간을 요하는지
- 복잡도 : 효율성에 반비례(복잡도가 높을수록 효율성 저하)

#### 시간 복잡도 분석
하드웨어 환경(부동소수처리 프로세서 존재 유무, 나눗셈 가속기능 유무, 입출력 장비의 성능, 공유여부), 소프트웨어 환경(프로그램 언어의 종류, 운영체제 및 컴파일러의 종류)등에 따라 달라지기 때문에, 분석이 어렵다
- 점근적 표기를 통해 간단히 복잡도를 표현(입력 크기에 대한 함수)
- 입력 크기 n이 무한대로 커질 때의 복잡도를 간단히 표현하는 표기법

1. Big-O-Notation : 복잡도의 점근적 상한
    - 실행시간이 f(n)에 비례하는 알고리즘
2. Big-Omega-Notation : 복잡도의 점근적 하한
    - 실행시간이 최소 f(n)만큼 걸린다
3. Big-Theta-Notation : 복잡도의 평균
    - 빅-오와 빅-오메가의 표기가 같은 경우에 사용
    - f(n)과 동일한 증가율을 가진다
