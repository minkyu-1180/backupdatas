## 3. Bubble Sort

### 버블 정렬(Bubble Sort)
인접한 두 개의 원소를 비교하며 자리를 계속 교환하는 방식
- 시간 복잡도 : O(n^2)
#### 버블 정렬 과정
1. 첫 번째 원소부터 인접한 원소끼리 계속 자리를 교환하면서 맨 마지막 자리까지 이동한다
2. 한 단계가 끝나면 가장 큰 원소가 마지막 자리로 정렬된다
3. 위의 과정을 시작 원소의 요소를 증가시키며 진행한다

```python
# arr : 버블정렬 전 배열 / N : 해당 배열의 원소 수(= len(arr))
def bubble_sort(arr, N):
    # i : N-1, N-2, ,,, , 1
    for i in range(N-1, 1, -1):
        # j : 0, 1, ,,, , i-1
        for j in range(i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
```
    
    

````python
def bubble_sort(arr, N):
    for i in range(N-1, 0, -1): 
        for j in range(i):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]
    return arr

my_arr = [12, 43, 11, 20, 59, 71, 32]
n = len(my_arr)
print(bubble_sort(my_arr, n))
````

## 4. Counting Sort

### 카운팅 정렬(Counting Sort)
항목들의 순서를 결정하기 위해 집합에 각 항목이 몇 개씩 있는지 세는 작업을 통해 선형 시간에 정렬하는 알고리즘
- 시간 복잡도 : O(n + k)
    - k : 정수의 최대값
#### 카운팅 정렬의 제한사항
- 정수나 정수로 표현할 수 있는 자료에 대해서만 적용 가능
- 집합 내의 가장 큰 정수를 알아야 함
- 0 이상의 정수 n이 비교적 작을 때만 가능

```python
# 1단계 : 개수 저장
data = [0, 4, 1, 3, 1, 2, 4, 1]
# 배열의 최대값 찾기(근데 보통 범위가 주어짐)
max = 0
for x in data:
    if x >= max:
        max = x
counts = [0] * (max + 1) # 4번 index까지 사용해야 해서
for x in data:
    for i in range(max + 1):
        if i == x:
            counts[i] += 1
# counts = [1, 3, 1, 1, 2]

# 2단계 : 누적
for i in range(1, len(counts))):
    counts[i] += counts[i-1]
# counts = [1, 4, 5, 6, 8]

# 3단계 : 정렬
temp = [0] * len(data)
for i in range(len(data)-1, -1, -1): # 역순
    # data[i]는 무조건 counts의 인덱스 번호 중 하나에 해당한다.
    # counts[data[i]]는 temp_idx 즉, 우리가 데이터를 넣을 index번호(ex. i = 7 -> data[i] = 1, counts[data[i]] = 4)
    temp_idx = counts[data[i]] 

    # temp의 해당 인덱스에 data의 요소를 삽입 후, counts에서 개수를 하나 빼줌
    temp[temp_idx] = data[i]
    counts[data[i]] -= 1
print(temp)
'''

````python
def Counting_Sort(A, k):
    # A : 원본 배열 
    # 입력 배열이 0 ~ k까지 즉, 데이터의 요소 범위까지
    
    counts = [0] * (k + 1)
    B = [0] * (k + 1)
    for i in range(0, len(A)):
        counts[A[i]] += 1
    for i in range(1, len(counts)):
        counts[i] += counts[i-1]
    for i in range(len(B)-1, -1, -1):
        counts[A[i]] -= 1
        B[counts[A[i]] = A[i]
    return B
    
````

## 6. Greedy Algorithm

### 탐욕 알고리즘(Greedy Algorithm)
최적해를 구하는 데 사용되는 근시안적인 방법
- 결정 할 때 마다 최적을 선택하여 최정 해답에 도달하는 방법
- 각 선택의 시점에서 이루어지는 결정은 지역적으로는 최적이지만, 최종 해답이 최적이라는 보장은 없다
- 머릿속에 떠오르는 생각을 검증 없이 바로 구현하면 Greedy 접근이 가능

#### 탐욕 알고리즘의 동작 과정
1. 해 선택 : 현재 상태에서 부분 문제의 최적 해를 구한 뒤, 이를 부분해 집합(Solution Set)에 추가
2. 실행 가능성 검사 : 새로운 부분해 집합이 실행 가능한지를 확인하고, 문제의 제약 조건을 위반하지 않는지 검사
3. 해 검사 : 새로운 부분해 집합이 문제의 해가 되는지 확인
   - 아직 전체 문제의 해가 완성되지 않았을 경우 처음부터 다시 시작

````python
# 6자리 주어지는 수, 세어주는 리스트
num = int(input())
# 그냥 error 방지를 위해 12자리까지 생성해놓기
# 밑의 while문에서 i가 9일 때, i + 2가 11이라서 11th index의 값에도 접근해야 됨..
counts = [0] * 12

for i in range(6):
    counts[num % 10] += 1
    num //= 10

# 시작 인덱스
i = 0 
# triplet, run 횟수 초기화
run = tri = 0

# 종료 조건 : 모든 인덱스를 다 둘러봄
while i < 10:
    # triplet 조건 성립
    if counts[i] >= 3:
        counts[i] -= 3
        tri += 1
        # if문에서 continue를 활용하여 밑의 if문도 확인하는 법(아니면 triplet이 될 경우 다시 while문으로 올라감)
        continue
    # run 조건 성립
    if counts[i] >= 1 and counts[i+1] >= 1 and counts[i+2] >= 1: 
        counts[i] -= 1
        counts[i+1] -= 1
        counts[i+2] -= 1
        run += 1
        continue
    i += 1
    
if run + tri == 2:
    print('Baby gin')
else:
    print('Lose')
````
