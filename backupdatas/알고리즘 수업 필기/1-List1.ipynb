````python
def Counting_Sort(A, k):
    # A :   
    #   0 ~ k ,   
    
    counts = [0] * (k + 1)
    B = [0] * (k + 1)
    for i in range(0, len(A)):
        counts[A[i]] += 1
    for i in range(1, len(counts)):
        counts[i] += counts[i-1]
    for i in range(len(B)-1, -1, -1):
        counts[A[i]] -= 1
        B[counts[A[i]] = A[i]
    return B
    
````

## 6. Greedy Algorithm

###  (Greedy Algorithm)
     
-          
-       ,     
-        Greedy  

####    
1.   :        ,   (Solution Set) 
2.    :      ,      
3.   :       
   -          

````python
# 6  ,  
num = int(input())
#  error   12 
#  while i 9 , i + 2 11 11th index   ..
counts = [0] * 12

for i in range(6):
    counts[num % 10] += 1
    num //= 10

#  
i = 0 
# triplet, run  
run = tri = 0

#   :    
while i < 10:
    # triplet  
    if counts[i] >= 3:
        counts[i] -= 3
        tri += 1
        # if continue   if  ( triplet    while )
        continue
    # run  
    if counts[i] >= 1 and counts[i+1] >= 1 and counts[i+2] >= 1: 
        counts[i] -= 1
        counts[i+1] -= 1
        counts[i+2] -= 1
        run += 1
        continue
    i += 1
    
if run + tri == 2:
    print('Baby gin')
else:
    print('Lose')
````
